```python
import numpy as np
import pandas as pd
from abc_analysis import abc_analysis
```

```python
df = pd.read_excel('Контейнерные перевозки (тонны).xlsx', sheet_name="Лист2")
df = df.drop(index = 0)
df
```

```python
abc = abc_analysis(df['Рынок итого'], True)
abc
```

```python
a_index = abc['Aind']
b_index = abc['Bind']
c_index = abc['Cind']

cond_list = [df.index.isin(a_index),
             df.index.isin(b_index),
             df.index.isin(c_index)]

choose_list = ['A', 'B', 'C']

df['abc'] = np.select(cond_list, choose_list)
df.sort_values(by=['Рынок итого'], ascending=False)
```
